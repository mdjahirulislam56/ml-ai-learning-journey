## Module 1 (Machine Learning)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'Salary' : [45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]})
df
```

```python
mn = df.min()
mx = df.max()
rng = mx - mn
print(rng)
```

```python
df_mean = df['Salary'].mean()
df_mean
```

```python

sq_df = (df['Salary'] - df_mean)**2
sum = sq_df.sum()
var = sum/(len(df)-1)
print(var)
df.var()
```

```python
np.var(df['Salary'], ddof=1) 
```

```python
std = np.sqrt(var)
print(std)
df.std(),np.std(df['Salary'],ddof=1)
```

```python
q2 = df.quantile(0.50)
print(q2)
```

```python
q3 = df.quantile(0.75)
print(q3)
```

```python
q1 = np.percentile(df,25)
q2 = np.percentile(df,50)
q3 = np.percentile(df,75)
q1,q2,q3
```

```python
IQR = q3 - q2
print(IQR)
```

```python
lower_fence = q1 - 1.5*IQR
upper_fence = q3 + 1.5*IQR
lower_fence,upper_fence
```

```python
outliers = df[((df['Salary']<lower_fence)|(df['Salary']>upper_fence))]
outliers
```
