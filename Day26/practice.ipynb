```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PracticeDay.csv')
df
```

```python
df['Discounted_Price'] = df['Price']-df['Price']*(10/100)

df
```

```python
df.duplicated(subset=['EmployeeID']).sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
print(df['Age'].isnull().sum())
print(df['Salary'].isnull().sum())
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df
```

```python
df.dropna(subset=['Salary'], inplace=True)
df
```

```python
print(df['Marks1'].mean())
print(df['Marks1'].max())
print(df['Marks2'].mean())
print(df['Marks2'].max())
print(df['Marks3'].mean())
print(df['Marks3'].max())
```

```python
df['Total Marks']= df['Marks1'] + df['Marks2'] + df['Marks3']
df
```

```python
df['Total Marks'].std()
```

```python
plt.hist(df['Age'],edgecolor='k')
plt.show()
```

```python
plt.scatter(df['Age'],df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(df['Price'],df['Discounted_Price'])
plt.xlabel('Price')
plt.ylabel('Discounted Price')
plt.show()
```
