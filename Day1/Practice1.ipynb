```python
num = eval(input('enter a num:'))
fact = 1
if num == 0:
  print(1)
else:
  while num>1:
    fact = fact * num
    num -= 1
  print(fact)
```

```python
num = input('enter a num:')
L = len(num)
sum = 0
for i in num:
  sum = sum + int(i)**L
print("Armstrong Number") if sum==int(num) else print("Not Armstrong Number")
```
