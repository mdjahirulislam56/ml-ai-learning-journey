```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_IQdata.csv")
df
```

```python
plt.scatter(df['Study_Hour'],df['IQ_Score'])
plt.xlabel('Study Hour')
plt.ylabel('IQ')
plt.show()
```

```python
plt.scatter(df['Shoe_Size'],df['Study_Hour'])
plt.xlabel('Shoe_Size')
plt.ylabel('Study_Hour')
plt.show()
```

```python
plt.hist(df["Shoe_Size"],bins=20,edgecolor='k')
plt.title('Histogram of Shoe_Size')
plt.xlabel('Shoe_Size')
plt.ylabel('Study_Hour')
plt.show()
```
