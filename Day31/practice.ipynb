### Part B


```python
total_mails = 1000

S = 100
F = 40
notS_hasF = 60

P_S = S/total_mails
P_F = (F+notS_hasF)/total_mails
P_FIS = F/S
P_SIF = (P_FIS*P_S) / P_F

print(P_SIF)
```

For a spam filter, the result \(P(S|F)=0.4\) means that if an email contains the word "free", there is a 40% chance that it is spam.
