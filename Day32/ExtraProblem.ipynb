### Problem 1

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'Weight_kg' : [58, 62, 65, 66, 190], 'Height_cm' : [150, 160, 170, 175, 180]})
df
```

```python
z_score = (df-df.mean())/df.std()

print(z_score)
```

```python
min_max = (df-df.min())/(df.max()-df.min())

print(min_max)
```

```python
robust_Scale = (df-df.median())/(df.quantile(0.75) - df.quantile(0.25))
print(robust_Scale)
```

```python
robust_Scale_weight = (df['Weight_kg']-df['Weight_kg'].median())/(df['Weight_kg'].quantile(0.75) - df['Weight_kg'].quantile(0.25))
print(robust_Scale_weight)
```

```python
least_fence = df['Weight_kg'].quantile(0.25)-1.5*(df['Weight_kg'].quantile(0.75)-df['Weight_kg'].quantile(0.25))
highest_fence = df['Weight_kg'].quantile(0.75)+1.5*(df['Weight_kg'].quantile(0.75)-df['Weight_kg'].quantile(0.25))
outlier_weight = df['Weight_kg'].loc[(df['Weight_kg']<least_fence)|(df['Weight_kg']>highest_fence)]
outlier_weight
```

### Problem 3

```python
import pandas as pd

data = {
    "ID": [1, 2, 3, 4],
    "City": ["Dhaka", "Chattogram", "Rajshahi", "Dhaka"],
    "Education_Level": ["High School", "Bachelor", "Master", "Bachelor"],
    "Satisfaction": ["Low", "Medium", "High", "Medium"]
}

df = pd.DataFrame(data)
print(df)
```

```python
city = pd.get_dummies(data['City'],prefix='',dtype=int)
city
```

```python
city_encoded = pd.concat([df,city],axis = 1)
city_encoded
```

```python
order = {'High School':0, 'Bachelor':1, 'Master':2}

educ= df['Education_Level'].map(order).astype(int)
educ
```

```python
educ_encoded = pd.concat([df,educ],axis=1)
educ_encoded
```

### Problem 5


- The **Euclidean distance** increases **more sharply** than the Manhattan distance when one coordinate changes dramatically, because it squares the differences.
- After scaling the **y-coordinate by 10**, both distances increased, but **Euclidean** grew **disproportionately**.
- This demonstrates why **feature scaling or normalization** is essential in distance-based models (e.g., k-NN, clustering):
  - Features with large numeric ranges can **dominate** the distance measure.
  - Proper scaling ensures **balanced influence** from all features.
