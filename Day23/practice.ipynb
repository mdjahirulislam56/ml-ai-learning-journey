## Part 1

```python
import numpy as np

score = np.random.randint(0,100,10)
print("Highest Score:",score.max())
print("Lowest Score:",score.min())
print("Average Score:",score.mean())
```

```python
arr = np.array([45, 60, 55, 30, 90, 75, 40, 50])

arr_more50 = arr[arr>50]
sorted_arr = np.sort(arr)
index = np.argmax(arr)

print(arr_more50,sorted_arr,index)
```

```python
temp = np.random.randint(20,40,(4,4))
print(temp)
avg = temp.mean(axis=1)
print(avg)
highest_t = temp.max()
print(highest_t)
temp[temp>35] = 35
print(temp)
```

```python
car_rec = np.random.randint(0,50,(3,5))
print(car_rec)

intersect = car_rec.sum(axis=0)
car_per_time = car_rec.sum(axis=1)

print("Total car per intersection: ",intersect,"\nTotal car per time slot:",car_per_time)
```

```python
stock = np.array([12, 15, 9, 20, 18])

remain = stock - stock*(10/100)

print(remain)
```
