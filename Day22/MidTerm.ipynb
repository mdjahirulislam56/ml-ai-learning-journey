```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

## Ans to the Q1

```python
age_array = np.array(ages)
salary_array = np.array(salaries)

print("age_array:",age_array)
print("salary_array:",salary_array)

print()

print(f"dtype of age_array: {age_array.dtype}")
print(f"ndim of age_array: {age_array.ndim}")
print(f"shape of age_array: {age_array.shape}")
print(f"size of age_array: {age_array.size}")

print()

print(f"dtype of salary_array: {salary_array.dtype}")
print(f"ndim of salary_array: {salary_array.ndim}")
print(f"shape of salary_array: {salary_array.shape}")
print(f"size of salary_array: {salary_array.size}")
```

## Ans to the Q2

```python
print("Highest salary value:",np.max(salary_array))
print("Lowest salary value:",np.min(salary_array))
print("Average salary:",np.average(salary_array))
print("Average age:",np.average(age_array))
```

## Ans to the Q3

```python
age_greater_30 = age_array[age_array>30]
print("Ages greater than 30:",age_greater_30)

older_than_30 = age_greater_30.size
print("No. of employees older than 30 yrs:",older_than_30)
```

```python
sum_of_salary = np.sum(salary_array)
print("Total salary expense:",sum_of_salary)

difference = np.max(salary_array) - np.min(salary_array)
print("Difference between the maximum and minimum salary:",difference)
```
