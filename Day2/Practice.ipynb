**Problem 02:** Write the function count_Letter(word, letter) which takes a word and a
letter as arguments and returns the number of occurrences of that letter in the word.

```python
def count_Letter(word, letter):
  count=0
  for i in word:
    if i == letter:
      count+=1
  return count
count_Letter('Bangladesh','a')
```

**Problem 07:** Given a route containing 4 directions (E, W, N, S). Find the shortest
path to reach destination.

```python
import math
def shortpath(string_path):
  x,y = 0,0
  for i in string_path:
    if i == "E":
      x+=1
    elif i == "W":
      x-=1
    elif i == "N":
      y+=1
    elif i == "S":
      y-=1
  short_dist = math.sqrt(x**2 + y**2)
  return short_dist

print (shortpath("WNEENESENNNEEEWWWNNNNSSS"))

```

**Problem 08:** Write a program to find substring from a string

```python
def substring(string,start,end):
  sub_s = ""
  for i in range(len(string)):
    print(i,end='')
    print(string[i],end=' ')
    if i >= start and i<= end:
      sub_s += string[i]
  return sub_s

substring("My name is Jahirul Islam",6,12)
```
